# Affine transformations and Homogeneous coordinates

$$P' = P*M + t$$

$P_1 = (0,0)$  
$P_2 = (2,0)$  
$P_3 = (2,1)$  
$P_4 = (1,2)$  
$P_5 = (0,1)$  


$$\begin{pmatrix}x' & y' \end{pmatrix}  = \begin{pmatrix}x & y \end{pmatrix} \begin{pmatrix} cos\theta & sin\theta \\ -sin\theta & cos\theta \end{pmatrix} + \begin{pmatrix}t_x & t_y \end{pmatrix}$$

$$\begin{pmatrix} 0 & 0 \\ 2 & 0 \\ 2& 1 \\ 1 & 2 \\ 0 & 1\end{pmatrix} * \begin{pmatrix} 1 & 0 \\ 0& -1 \end{pmatrix} + \begin{pmatrix} 0 & 0\end{pmatrix} = \begin{pmatrix} 0 & 0 \\ 2 & 0 \\ 2& 2 \\ 1 & 4 \\ 0 & 2\end{pmatrix}$$

Our transformation can be written as:  
$$ \begin{pmatrix}0 & 1 & 0 \\ -1 & 0 & 0 \\ 1 & 1 & 1\end{pmatrix}$$

Cartesian to homogeneous: $$\begin{pmatrix} x & y \end{pmatrix} = \begin{pmatrix} x & y & 1\end{pmatrix}$$

$$\begin{pmatrix} 0 & 0 & 1 \\ 2 & 0 & 1  \\ 2& 1& 1  \\ 1 & 2& 1  \\ 0 & 1& 1 \end{pmatrix} * \begin{pmatrix} 1 & 0 & 0\\ 0& -1 & 0 \\ 1 & 1 &  1\end{pmatrix} = \begin{pmatrix} 0 & 0 & 1 \\ 2 & 0 & 1 \\ 2& 2 & 1 \\ 1 & 4& 1  \\ 0 & 2& 1 \end{pmatrix}$$

Just add ones and remove ones at the end

$$\begin{pmatrix} 0 & 0 & 1 \\ 2 & 0 & 1  \\ 2& 1& 1  \\ 1 & 2& 1  \\ 0 & 1& 1 \end{pmatrix} * \begin{pmatrix} 1 & 0 & 0.2\\ 0& 1 & 0.1 \\ 1 & 1 &  1\end{pmatrix} = \begin{pmatrix} 0 & 0 & 1 \\ 2 & 0 & 1.4 \\ 2& 2 & 1.5 \\ 1 & 4& 1.4  \\ 0 & 2& 1.1 \end{pmatrix}$$

In this case we can't just remove the ones in the last columns since they're not 1s, we first have to divide like $(x y z) = (\frac{x}{z}\frac{y}{z}\frac{z}{z}) = (\frac{x}{z}\frac{y}{z}1)$

$$\begin{pmatrix} 0 & 0 & 1 \\ 2 & 0 & 1.4 \\ 2& 2 & 1.5 \\ 1 & 4& 1.4  \\ 0 & 2& 1.1 \end{pmatrix} = \begin{pmatrix} 0 & 0 & 1\\ 1.43 & 0 & 1 \\ 1.33& 2 & 0.67 \\ 0.71 & 1.43& 1  \\ 0 & 0.91& 1 \end{pmatrix}$$
