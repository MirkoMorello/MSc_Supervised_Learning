# Key ideas of CNNs:
- Feed forward feature extraction
- supervised training of convolutional filters by back-propagating calssification error
- with respect to traditional NN, CNNs have other special layers:
  - Spatial pooling
  - Local response nornmalization
- These layers are useful to:
  - Reduce Computational burden
  - increase variance
  - ease the optimization

## Local Response Normalization

In order to, for example, avoid learning the contrast because I'm just interested in learning the content of an image, I can use LRP, we have two types:
- Within channel
- Across channel, just considering one spatial direction and performing normalization across that channel

## Summary:

- Two key differences to Standard Neural Nets: Neurons arranged in 3D volumes have local connectivity, share weights
