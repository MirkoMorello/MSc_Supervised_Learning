## Boosting
first ensemble method we're about to see, it refers to a family of algorithms that can convert weak learners to strong learners.  
weak learner is just slightly better than random guess, while a strong learner is very close to perfect performance.  
It's much easier to take a weak learner and boost it than to create a strong one.  

Boosting works by training a set of learners sequentially and combining them for rediction, where the later learners focus more on the mistakes of the earlier learners.

## Random Tree Ensembles (Random Forests)

it's the state of the art ensemble method.  
it is an extension of bagging, where the major difference with Bagging is the incorporation of randomized feature selection

The parameter $K$ controls the incorporation of randomness.  
When $K$ equals the total number of features, the constructed decision tree is identical to the traditional deterministic decision tree, when $K=1$ a feature will be selected randomly,
 the suggested value of $K$ is the logarithm of the number of features.


The convergence property of RF is similar to that of Bagging, but it starts with an higher test error and ends up with a lower one thanks to its properties.
